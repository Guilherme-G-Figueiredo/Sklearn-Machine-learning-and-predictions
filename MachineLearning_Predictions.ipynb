# Python Project AI: Artificial Intelligence and Predictions
### Case: Clients' Credit Score
You have been hired by a bank to determine the credit score of clients. You need to analyze all the bank's clients and, based on this analysis, create a model that can read the client's information and automatically determine their credit score: Poor, Fair, Good.

Class files: https://drive.google.com/drive/folders/1FbDqVq4XLvU85VBlVIMJ73p9oOu6u2-J?usp=drive_link

```python
import pandas as pd

data = pd.read_csv("Clients.csv")
display(data)
display(data.info())
```

```python
#profession, credit_mix e payment_behavior need label encoding
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

data["profession"] = encoder.fit_transform(data["profession"])
data["credit_mix"] = encoder.fit_transform(data["credit_mix"])
data["payment_behavior"] = encoder.fit_transform(data["payment_behavior"])
display(data)
display(data.info())
```

```python
from sklearn.model_selection import train_test_split

y = data["credit_score"]
x = data.drop(columns=["credit_score", "client_id"])

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

RandomForest_model = RandomForestClassifier()
KNN_model = KNeighborsClassifier()

RandomForest_model = RandomForest_model.fit(x_train, y_train)
KNN_model = KNN_model.fit(x_train, y_train)

previsao_arvoreDecisao = RandomForest_model.predict(x_test)
previsao_knn = KNN_model.predict(x_test)

from sklearn.metrics import accuracy_score

display(accuracy_score(y_test, previsao_arvoreDecisao))
display(accuracy_score(y_test, previsao_knn))
```
